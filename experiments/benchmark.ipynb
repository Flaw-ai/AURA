```python
import time
import torch

from modeling.modeling_flaw import (
    FlawForCausalLM
)

from modeling.generation_flaw import (
    FlawGenerator
)
```

```python
model = FlawForCausalLM.from_pretrained(
    "../models/flaw"
)

generator = FlawGenerator(
    model,
    tokenizer
)
```

```python
prompt = """
Explain Newton's Second Law.
"""

start = time.time()

response = generator.generate(
    prompt
)

end = time.time()

print(
    "Latency:",
    end - start
)
```

```python
tokens = len(
    tokenizer.encode(
        response
    )
)

speed = tokens / (
    end - start
)

print(
    "Tokens/sec:",
    speed
)
```

```python
params = sum(
    p.numel()
    for p in model.parameters()
)

print(
    "Parameters:",
    params
)

print(
    "Model Size MB:",
    params * 4 / 1024 / 1024
)
```
