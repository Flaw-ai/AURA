```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(
    "../models/flaw"
)

tests = [

    "Hello world",

    "नमस्ते दुनिया",

    "What is photosynthesis?",

    "Solve x² + 5x + 6 = 0",

    "The mitochondria is the powerhouse of the cell.",

    "<|user|> Hello",

    "<|assistant|> Hi"
]

for text in tests:

    ids = tokenizer.encode(text)

    decoded = tokenizer.decode(ids)

    print("=" * 60)

    print("TEXT:")
    print(text)

    print("\nTOKENS:")
    print(len(ids))

    print(ids)

    print("\nDECODED:")
    print(decoded)
```

```python
samples = [

    "English sentence",

    "Hindi sentence",

    "Hinglish sentence",

    "Math equation",

    "NCERT paragraph"
]

for sample in samples:

    token_count = len(
        tokenizer.encode(sample)
    )

    chars = len(sample)

    print(
        sample,
        chars,
        token_count,
        chars / token_count
    )
```
